```python
def split_text_into_chunks(text, max_tokens):
    """Splits text into chunks of a maximum number of tokens."""
    chunks = []
    current_chunk = []
    current_token_count = 0

    for word in text.split():
        current_chunk.append(word)
        current_token_count += 1

        if current_token_count >= max_tokens:
            chunks.append(" ".join(current_chunk))
            current_chunk = []
            current_token_count = 0

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks
```

```python
import vertexai
from vertexai.language_models import TextEmbeddingInput, TextEmbeddingModel
import json
import pandas as pd

# Replace with your project ID, region, and bucket name
project_id = "rv-hermes-nonprod"
region = "us-central1"

# Initialize Vertex AI
vertexai.init(project=project_id, location=region)

model = TextEmbeddingModel.from_pretrained("text-embedding-004")

# Load your JSON data from a local file
with open("monstro_TechProduct_11_09_2024.json", "r") as f:
    products = json.load(f)

# Create a list to store embeddings and document IDs
embeddings_data = []

# Batch size for embedding generation
batch_size = 10

# Iterate through the products in batches
for i in range(0, len(products), batch_size):
    batch_products = products[i:i + batch_size]

    # Create a list of TextEmbeddingInput objects for the batch
    text_embedding_inputs = [
        TextEmbeddingInput(task_type="RETRIEVAL_DOCUMENT", text=convert_json_to_string(product))
        for product in batch_products
    ]

    try:
        # Generate embeddings for the batch
        embeddings = model.get_embeddings(text_embedding_inputs)

        # Append the embeddings to the list, keeping track of document IDs manually
        for j, embedding in enumerate(embeddings):
            embeddings_data.append(
                {"document_id": batch_products[j]["uuid"], "embedding": embedding.values}
            )

    except Exception as e:
        print(f"Error generating embeddings for batch {i}: {e}")

# Create a Pandas DataFrame from the embeddings data
embeddings_df = pd.DataFrame(embeddings_data)

# Save the embeddings to a CSV file (adjust the file name as needed)
embeddings_df.to_csv("product_embeddings.csv", index=False)
```

Convert embeddings to format for vector search

```python
import csv
import json

def convert_csv_to_jsonl(input_csv_file, output_jsonl_file):
    with open(input_csv_file, 'r', newline='', encoding='utf-8') as csv_file:
        reader = csv.DictReader(csv_file)

        with open(output_jsonl_file, 'w', encoding='utf-8')) as jsonl_file:
            for row in reader:
                document_id = row['document_id']
                # Convert the embedding string to a list
                embedding = eval(row['embedding'])  # Use eval carefully; ensure the input is safe
                
                # Create a new dictionary for the desired format
                json_object = {
                    "id": document_id,
                    "embedding": embedding
                }
                
                # Write the JSON object as a line in the JSONL file
                jsonl_file.write(json.dumps(json_object) + '\n')

# Example usage
input_csv_file = 'product_embeddings_11_09_2024.csv'  # Your input CSV file
output_jsonl_file = 'vector_search_embeddings_format_11_09_2024.jsonl'  # Desired output JSONL file

convert_csv_to_jsonl(input_csv_file, output_jsonl_file)
```
