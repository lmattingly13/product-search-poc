```python
import pandas as pd

df = pd.read_json('monstro_TechProduct_11_09_2024.json')
df.head()
```

```python
df['product_name'] = df['product'].apply(lambda x: x['name'])
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample Text Data
corpus = df.product_name.tolist()

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and Transform
vectorizer.fit_transform(corpus)
```

```python
def get_sparse_embedding(text):
    # Transform Text into TF-IDF Sparse Vector
    tfidf_vector = vectorizer.transform([text])

    # Create Sparse Embedding for the New Text
    values = []
    dims = []
    for i, tfidf_value in enumerate(tfidf_vector.data):
        values.append(float(tfidf_value))
        dims.append(int(tfidf_vector.indices[i]))
    return {"values": values, "dimensions": dims}
```

```python
text_test = "Tracfone - TCL 30 Z 32GB Prepaid with 1 Year of Service Bundle - Black"
get_sparse_embedding(text_test)
```

```python
items = []
for i in range(len(df)):
    id = df.uuid[i]
    product_name = df.product_name[i]
    sparse_embedding = get_sparse_embedding(product_name)
    items.append({"id": id, "product_name": product_name, "sparse_embedding": sparse_embedding})
items[:5]
```

```python
with open("sparse_embedding_products_11_09_2024.json", "w") as f:
    for item in items:
        f.write(f"{item}\n")
```

Create a spare embedding index in vector search

```python
from google.cloud import aiplatform
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/Users/lmattingly/Downloads/rv-hermes-nonprod-3a8067e91511.json"

# Replace with your project ID, region, and bucket name
project_id = "rv-hermes-nonprod"
region = "us-central1"

aiplatform.init(project=project_id, location=region)
```

```python
BUCKET_URI = f"gs://commerce-product-search-data/sparse_embedding_products_11_09_2024.json"

UID= "sparse_embedding_products_11_09_2024"

# create Index (this should finish less than one minute)
my_sparse_index = aiplatform.MatchingEngineIndex.create_tree_ah_index(
    display_name=f"commerce-product-search-index-poc_{UID}",
    contents_delta_uri=BUCKET_URI,
    dimensions=768,
    approximate_neighbors_count=10,
)
```

```python
ENDPOINT = "commerce-product-index-endpoint-poc_{UID}"

my_index_endpoint = aiplatform.MatchingEngineIndexEndpoint.create(
    display_name=ENDPOINT, public_endpoint_enabled=True
)
```

```python
DEPLOYED_TOKEN_INDEX_ID = f"commerce_product_search_deployed_{UID}"
my_index_endpoint.deploy_index(
    index=my_sparse_index, deployed_index_id=DEPLOYED_TOKEN_INDEX_ID
)
```

```python
from google.cloud.aiplatform.matching_engine.matching_engine_index_endpoint import (
    HybridQuery,
)

# create HybridQuery
query_text = "Tvs"
query_emb = get_sparse_embedding(query_text)
query = HybridQuery(
    sparse_embedding_dimensions=query_emb["dimensions"],
    sparse_embedding_values=query_emb["values"],
)
```

```python
ENDPOINT_ID = "1809721372528607232"
my_index_endpoint = aiplatform.MatchingEngineIndexEndpoint(
    index_endpoint_name=f"{ENDPOINT_ID}"
)
```

```python

DEPLOYED_TOKEN_INDEX_ID = '6569938217728999424'
# build a query request
response = my_index_endpoint.find_neighbors(
    deployed_index_id=DEPLOYED_TOKEN_INDEX_ID,
    queries=[query],
    num_neighbors=5,
)

# print results
for idx, neighbor in enumerate(response[0]):
    print(f"Neighbor {idx+1}, nei")
    title = df.title[int(neighbor.id)]
    print(f"{title:<40}")
```
