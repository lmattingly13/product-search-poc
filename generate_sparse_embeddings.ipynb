```python
import pandas as pd

df = pd.read_json('monstro_TechProduct_11_09_2024.json')
df.head()
```

```python
df['product_name'] = df['product'].apply(lambda x: x['name'])
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample Text Data
corpus = df.product_name.tolist()

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and Transform
vectorizer.fit_transform(corpus)
```

```python
def get_sparse_embedding(text):
    # Transform Text into TF-IDF Sparse Vector
    tfidf_vector = vectorizer.transform([text])

    # Create Sparse Embedding for the New Text
    values = []
    dims = []
    for i, tfidf_value in enumerate(tfidf_vector.data):
        values.append(float(tfidf_value))
        dims.append(int(tfidf_vector.indices[i]))
    return {"values": values, "dimensions": dims}
```

```python
text_test = "Tracfone - TCL 30 Z 32GB Prepaid with 1 Year of Service Bundle - Black"
get_sparse_embedding(text_test)
```

```python
items = []
for i in range(len(df)):
    id = df.uuid[i]
    product_name = df.product_name[i]
    sparse_embedding = get_sparse_embedding(product_name)
    items.append({"id": id, "product_name": product_name, "sparse_embedding": sparse_embedding})
items[:5]
```

```python
import json
import re

def clean_json_data(raw_records):
    cleaned_records = []
    
    for record in raw_records:
        # Convert the record to a string if it is not already
        record_str = json.dumps(record) if isinstance(record, dict) else record
        
        # Replace single quotes with double quotes
        record_str = record_str.replace("'", '"')

        # Escape inner double quotes
        record_str = re.sub(r'(?<!\\)"', r'\\"', record_str)  # Escape unescaped double quotes

        # Fix escape sequences
        record_str = re.sub(r'\\xa0', ' ', record_str)  # Replace \xa0 with a space
        record_str = re.sub(r'\\', '', record_str)  # Remove any other unnecessary backslashes
        
        # Load the JSON to validate it
        try:
            cleaned_record = json.loads(record_str)
            cleaned_records.append(cleaned_record)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
            print(f"Record that failed: {record_str}")
    
    return cleaned_records
```

```python
with open("sparse_embedding_products_11_09_2024.json", "r") as input_file:
    raw_records = input_file.readlines()
```

```python

# Clean the raw records
cleaned_data = clean_json_data(raw_records)

# Save the cleaned data to a new JSON file with a 'cleaned_' prefix
output_file_path = 'cleaned_sparse_embedding_products_11_09_2024.json'
with open(output_file_path, 'w') as output_file:
    json.dump(cleaned_data, output_file, indent=2)
```

```python
from google.cloud.aiplatform.matching_engine.matching_engine_index_endpoint import (
    HybridQuery,
)

# create HybridQuery
query_text = "Tvs"
query_emb = get_sparse_embedding(query_text)
query = HybridQuery(
    sparse_embedding_dimensions=query_emb["dimensions"],
    sparse_embedding_values=query_emb["values"],
)
```

```python
ENDPOINT_ID = "3582310038363963392"
my_index_endpoint = aiplatform.MatchingEngineIndexEndpoint(
    index_endpoint_name=f"{ENDPOINT_ID}"
)
```

```python

DEPLOYED_TOKEN_INDEX_ID = 'commerce_product_search_deployed_sparse_embedding_products_11_09_2024'
# build a query request
response = my_index_endpoint.find_neighbors(
    deployed_index_id=DEPLOYED_TOKEN_INDEX_ID,
    queries=[query],
    num_neighbors=5,
)

# print results
for idx, neighbor in enumerate(response[0]):
    print(f"Neighbor {idx+1}, nei")
    title = df.title[int(neighbor.id)]
    print(f"{title:<40}")
```
